```python
import numpy as np
import cv2 # computer vison
import matplotlib.pyplot as plt
%matplotlib inline 
#dung cho viec hien thi tren cung 1 file
```

```python
img = cv2.imread('data/rudi.jpg')
#convet gray
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
print(gray.shape)
plt.imshow(gray,cmap='gray')
plt.show
```

```python
#shrink
#w = 100, h = 50
img_re = cv2.resize(gray,(100,50),cv2.INTER_AREA)
img_re.shape
```

```python
plt.imshow(img_re,cmap='gray')
plt.show()
```

```python
#englarge
#w= 1000, h=500
img_en = cv2.resize(gray,(1000,500),cv2.INTER_CUBIC)
img_en.shape
```

```python
plt.imshow(img_en,cmap='gray')
plt.show()
```
